## Probability Distribution Functions

```python
# for inline plots%matplotlib inline
import matplotlib.pyplot as plt

# import seaborn
import seaborn as sns
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(4.5,3)})

from random import seed
```

### Continuous

#### Generate random numbers following  Uniform distribution

```python
# import uniform distribution
from scipy.stats import uniform

# generate uniform random numbers
seed(1)
n = 10000
a = 0
b = 10
data_uniform = uniform.rvs(size=n, loc = a, scale=b)
```

```python
ax = sns.distplot(data_uniform,
                  bins=100)
ax.set(xlabel='Uniform ', ylabel='Frequency')
```

#### Generate random numbers following Exponential distribution

```python
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)

ax = sns.distplot(data_expon,
                  bins=100)
ax.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

#### Generate random numbers following Normal distribution

```python
from scipy.stats import norm
# generate random numbers from normal distribution
data_normal = norm.rvs(size=10000,loc=0,scale=1)

ax = sns.distplot(data_normal,
                  bins=100)
ax.set(xlabel='Normal', ylabel='Frequency')
```
